#knn classifer

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
x=df[['Weight', 'Height']]
y= df[['Gender']]
```

```python
knn_cls= KNeighborsClassifier(n_neighbors=3,weights='distance')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
knn_cls.fit(x_train,y_train)
```

```python
test_pred_c=knn_cls.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, test_pred_c)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
height = float(input("Enter height in inches: "))
weight = float(input("Enter weight in lbs: "))
manual_input = [[height, weight]]

predicted_gender = knn_cls.predict(manual_input)
gender_label = "Male" if predicted_gender[0] == 1 else "Female"
gender_label
```
